## Выборки и агрегации

### 1. Загрузите файл JSON, созданный в предыдущем упражнении, в фрейм данных
- установить CarNumber в качестве индексного столбца

```python
import pandas as pd

df = pd.read_json('../data/auto.json')
df.set_index('CarNumber', inplace=True)
print(df)
```

### 2. Сделайте следующий выбор
- отображать только те строки, где штрафы превышают 2100
- отображать только те строки, где штрафы больше 2100, а возврат равен 2
- отображать только те строки, где есть модели из списка:[’Focus’, ’Corolla’]
- отображать только те строки, где номер автомобиля есть из списка: [’Y7689C197RUS’, ’92928M178RUS’, ’7788KT197RUS’, ’H115YO163RUS’, ’X758HY197RUS’]

```python
fines_above_2100 = df[df['Fines'] > 2100]
fines_above_2100_return_2 = df[(df['Fines'] > 2100) & (df['Refund'] == 2)]
models_focus_corolla = df[df['Model'].isin(['Focus', 'Corolla'])]
car_numbers_list = ['Y7689C197RUS', '92928M178RUS', '7788KT197RUS', 'H115YO163RUS', 'X758HY197RUS']
specific_car_numbers = df.loc[car_numbers_list]
print(f"Штрафы превышают 2100: \n{fines_above_2100}")
print(f"\nШтрафы превышают 2100, а возврат равен двум: \n{fines_above_2100_return_2}")
print(f"\nТолько Фокусы и Короллы: \n{models_focus_corolla}")
print(f"\nОсобые номера авто: \n{specific_car_numbers}")
```

### 3. Сделать агрегации с маркой и моделью
- отобразить средние штрафы, сгруппированные по маркам
- отобразить средние штрафы, сгруппированные по марке и модели
- отобразить количество штрафов, сгруппированных по марке и модели, чтобы понять, можно ли доверять медианным значениям
- отображать минимальные и максимальные штрафы, сгруппированные по марке и модели, чтобы лучше понять разницу
- отображать стандартное отклонение штрафов, сгруппированных по марке и модели, чтобы лучше понять дисперсию

```python
average_fines_by_make = df.groupby('Make')['Fines'].mean() # Средние штрафы по маркам
average_fines_by_make_model = df.groupby(['Make', 'Model'])['Fines'].mean() # Средние штрафы по марке и модели
count_fines_by_make_model = df.groupby(['Make', 'Model']).size() # Количество штрафов по марке и модели
min_max_fines_by_make_model = df.groupby(['Make', 'Model'])['Fines'].agg(['min', 'max']) # Минимальные и максимальные штрафы по марке и модели
std_fines_by_make_model = df.groupby(['Make', 'Model'])['Fines'].std()

print(f"Средние штрафы по маркам: \n{average_fines_by_make}")
print(f"\nСредние штрафы по марке и модели: \n{average_fines_by_make_model}")
print(f"\nКоличество штрафов по марке и модели: \n{count_fines_by_make_model}")
print(f"\nМинимальные и максимальные штрафы по марке и модели: \n{min_max_fines_by_make_model}")
print(f"\nСтандартное отклонение штрафов по марке и модели: \n{std_fines_by_make_model}")
```

### 4. Сделать агрегации с номером автомобиля
- вывести номера автомобилей, сгруппированные по количеству штрафов в порядке убывания, мы хотим найти тех, кто чаще всего нарушал закон
- выберите из исходного фрейма данных все строки, соответствующие первому номеру автомобиля, мы хотим немного увеличить масштаб
- вывести номера автомобилей, сгруппированные по сумме штрафов в порядке убывания, мы хотим найти тех, кто заплатил больше всего
- выберите из исходного фрейма данных все строки, соответствующие первому номеру автомобиля, мы хотим немного увеличить масштаб
- отобразить таблицу, которая отвечает на вопрос: есть ли номера автомобилей, которые были связаны с разными моделями?

```python
car_numbers_by_fines_count = df.groupby('CarNumber').size().sort_values(ascending=False)

top_car_number = car_numbers_by_fines_count.index[0]
top_car_number_data = df.loc[top_car_number]

car_numbers_by_fines_sum = df.groupby('CarNumber')['Fines'].sum().sort_values(ascending=False)

top_car_number_by_sum = car_numbers_by_fines_sum.index[0]
top_car_number_data_by_sum = df.loc[top_car_number_by_sum]

car_numbers_with_multiple_models = df.groupby('CarNumber')['Model'].nunique()
car_numbers_with_multiple_models = car_numbers_with_multiple_models[car_numbers_with_multiple_models > 1]

print(f"Номера автомобилей, сгруппированные по количеству штрафов в порядке убывания: \n{car_numbers_by_fines_count}")
print(f"\nВыбор из исходного фрейма данных всех строк, соответствующих первому номеру автомобиля: \n{top_car_number_data}")
print(f"\nНомера автомобилей, сгруппированные по сумме штрафов в порядке убывания: \n{car_numbers_by_fines_sum}")
print(f"\nВыбор из исходного фрейма данных всех строк, соответствующих первому номеру автомобиля: \n{top_car_number_data_by_sum}")
print(f"\nОтображение таблицы: есть ли номера автомобилей, которые были связаны с разными моделями: \n{car_numbers_with_multiple_models}")
```
